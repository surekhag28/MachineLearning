```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df.head()
```

```python
df.columns
```

```python
df = df[['active', 'cured', 'death', 'new_active', 'new_cured', 'new_death',
       'new_positive', 'positive', 'state_code', 'state_name']]
df
```

```python
max(df['positive']) # maximum covid positive cases
```

```python
total_states = np.arange(len(df['state_name']))
total_states
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['positive'],color='g')
plt.yticks(total_states,df['state_name'])
plt.xlabel("Positive covid cases")
plt.title("Coronavirus cases")
```

```python
figure(num=None,figsize=(9,6),dpi=80,facecolor='w',edgecolor='k')
plt.barh(total_states,df['new_active'],color='y')
plt.yticks(total_states,df['state_name'])
plt.xlabel("Current active covid cases")
plt.title("Active coronavirus cases in each state")
```

```python
figure(num=None,figsize=(9,6),dpi=80,facecolor='w',edgecolor='k')
plt.barh(total_states,df['new_death'],color='y')
plt.yticks(total_states,df['state_name'])
plt.xlabel("Death covid cases")
plt.title("Death coronavirus cases in each state")
```
